```python
# needed librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


from scipy.stats import chi2_contingency
from scipy.stats import boxcox
import scipy.stats as stats

# classification model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv(r'./../data/df_clean.csv') 
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
numericals = df.select_dtypes(include=['float64', 'int64'])
numericals.head()

# We're making a new dataframe called "numericals" in which we take off all categoricals
```

```python
columns_to_drop = ['customer_number', 'bank_accounts_open', 'credit_cards_held', 'homes_owned', 'household_size']
numericals = numericals.drop(columns=columns_to_drop)

# and then we drop discrete variables so we only keep the continuous
```

```python
numericals.head()
```

```python
# Apply Box-Cox transformation to column
transformed_column, _ = boxcox(numericals['average_balance'])

# Replace original column with transformed column
numericals['average_balance'] = transformed_column
```

```python
numericals.head()
```

```python
# Get the data from the "average_balance" column of the "numericals" dataset
data = numericals["average_balance"]

# Set the style to "whitegrid" for a cleaner look
sns.set_style("whitegrid")

# Create a histogram of the data with a blue color and 20 bins
plt.hist(data, color="skyblue", bins=20)

# Add labels and title with a larger font size
plt.xlabel("Average Balance", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.title("Distribution of Average Balance", fontsize=16)

# Add a vertical line to show the mean value
plt.axvline(x=data.mean(), color="red", linestyle="--", linewidth=2, label="Mean")

# Add a legend for the mean value line
plt.legend()

# Show the plot
plt.show()
```

```python
# Find non-positive values in the 'q1_balance' column
mask = numericals['q1_balance'] <= 0

# Replace non-positive values with the mean of the positive values
positive_values = numericals.loc[~mask, 'q1_balance']
mean_value = np.mean(positive_values)
numericals.loc[mask, 'q1_balance'] = mean_value

# Apply Box-Cox transformation to column
transformed_column, _ = boxcox(numericals['q1_balance'])

# Replace original column with transformed column
numericals['q1_balance'] = transformed_column
```
